```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(healthcare)
```

```python
import numpy as np
healthcare['MonthlyIncome'] = np.log10(healthcare['MonthlyIncome'])

healthcare['Yearly_Increment'] = np.log10(healthcare['Yearly_Increment'])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(healthcare, healthcare["Attrition"]):
    strat_train_set = healthcare.loc[train_index]
    strat_test_set = healthcare.loc[test_index]
```

```python
strat_test_set["Attrition"].value_counts() / len(strat_test_set)
```

```python
healthcare["Attrition"].value_counts() / len(healthcare)
```

```python
strat_train_set.head().T
```

```python
healthcare = strat_train_set.drop('Attrition', axis = 1)
attrition = strat_train_set['Attrition']
test = strat_test_set.drop('Attrition', axis = 1)
y_test = strat_test_set['Attrition']
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(n_jobs = -1)
model.fit(healthcare, attrition)
y_pred = model.predict(test)
accuracy_score(y_pred,y_test)
```
